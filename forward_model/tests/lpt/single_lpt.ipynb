```python
%load_ext autoreload
%autoreload 2

import sys

import jax
import jax.numpy as jnp

sys.path.append('../../')
from ics import get_delta
from lpt import get_forward_lpt
from plot_utils import plot_cubes
```

```python
N, L, Z_I, Z_F = 128, 120, 99, 0

SEED_INT = 1
key = jax.random.PRNGKey(SEED_INT)
input_arr = jax.random.normal(key, shape=(N**3,))

INPUT_ARG = "GAUSSIAN_NCUBE_NUMBERS_FLAT"
LPT_METHOD = "ALPT"
R_S = 4
MUSCLE = True
MUSCLE_ITERS = "MANY"

forward_lpt = get_forward_lpt(
    N,
    L,
    Z_I,
    Z_F,
    LPT_METHOD=LPT_METHOD,
    INPUT_ARG=INPUT_ARG,
    MUSCLE=MUSCLE,
    MUSCLE_ITERS=MUSCLE_ITERS,
    R_S=R_S
)

delta_fin = forward_lpt(input_arr)
```

```python
fig, ax = plot_cubes(delta_fin, cmap='gnuplot', vlim=(-1, 10))
```
