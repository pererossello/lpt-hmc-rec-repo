```python
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
x = jnp.linspace(-10, 10, 1000)
C = 10

alpha = jnp.exp(-x)
alpha_ = jnp.exp(-x/C)

fig, ax = plt.subplots(1, 1)

ax.plot(x, alpha, label='not scaled')
ax.plot(x, alpha_, label='scaled')

ax.plot([x.min(), x.max()], [1, 1], ls='--', color='grey')

ylims = (0,1.5)
ax.plot([0, 0], [ylims[0], ylims[1]], ls='--', color='grey')

# ax.set_yscale('log')
ax.set_ylim(ylims[0], ylims[1])
ax.legend()
```
